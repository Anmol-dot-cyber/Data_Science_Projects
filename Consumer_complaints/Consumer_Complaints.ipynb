```python
# fill the na for outlet size with medium, here its replaced dynamically
train_data['State'] = train_data['State'].fillna(train_data['State'].mode()[0])
test_data['State'] = test_data['State'].fillna(test_data['State'].mode()[0])
```

```python
#One-hot encoding, creating unique row values as a column
for col in ['Tags','Company public response','Issue','State','Product','Company response to consumer','Consumer consent provided?','Submitted via','Sub-issue','Sub-product']:
    
    dummy=pd.get_dummies(train_data[col],prefix=col,drop_first=True)
    train_data=pd.concat([dummy,train_data],1)
    train_data.drop([col],1,inplace=True)
    
    dummy=pd.get_dummies(test_data[col],prefix=col,drop_first=True)
    test_data=pd.concat([dummy,test_data],1)
    test_data.drop([col],1,inplace=True)
```

# Decision tree

Importing required Liabraries

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

Building Dercision Tree Model

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
y_pred[:200]
```

```python
y_test[:200]
```

Evaluating Model

```python
# Model Accuracy, how often is the classifier correct?
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
```

```python
results = confusion_matrix(y_test, y_pred) 
print('Confusion Matrix :')
print(results)
print('Accuracy Score :',accuracy_score(y_test, y_pred))
print('Report : ')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
roc_auc_score(y_train, clf.predict(X_train))
```

```python
roc_auc_score(y_test, clf.predict(X_test))
```
