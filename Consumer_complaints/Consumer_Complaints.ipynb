```python
# fill the na for outlet size with medium, here its replaced dynamically
train_data['State'] = train_data['State'].fillna(train_data['State'].mode()[0])
test_data['State'] = test_data['State'].fillna(test_data['State'].mode()[0])
```

```python
#One-hot encoding, creating unique row values as a column
for col in ['Tags','Company public response','Issue','State','Product','Company response to consumer','Consumer consent provided?','Submitted via','Sub-issue','Sub-product']:
    
    dummy=pd.get_dummies(train_data[col],prefix=col,drop_first=True)
    train_data=pd.concat([dummy,train_data],1)
    train_data.drop([col],1,inplace=True)
    
    dummy=pd.get_dummies(test_data[col],prefix=col,drop_first=True)
    test_data=pd.concat([dummy,test_data],1)
    test_data.drop([col],1,inplace=True)
```

```python
#https://docs.python.org/3/library/datetime.html
#https://docs.python.org/2/library/datetime.html#datetime-objects

train_data['Date sent to company'] =  pd.to_datetime(train_data['Date sent to company'])
train_data['Date received'] =  pd.to_datetime(train_data['Date received'])
train_data['sent_receive_date_diff']=(train_data['Date sent to company']-train_data['Date received'])


test_data['Date sent to company'] =  pd.to_datetime(test_data['Date sent to company'])
test_data['Date received'] =  pd.to_datetime(test_data['Date received'])
test_data['sent_receive_date_diff']=(test_data['Date sent to company']-test_data['Date received'])
```

```python
train_data['sent_receive_date_diff']=pd.to_numeric(train_data['sent_receive_date_diff'])
train_data['sent_receive_date_diff']=train_data['sent_receive_date_diff']/86400000000000

test_data['sent_receive_date_diff']=pd.to_numeric(test_data['sent_receive_date_diff'])
test_data['sent_receive_date_diff']=test_data['sent_receive_date_diff']/86400000000000
```

```python
for col in ['Date received','Date sent to company']:
    train_data.drop([col],1,inplace=True)
    test_data.drop([col],1,inplace=True)
```
